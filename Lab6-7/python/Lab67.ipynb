```python
###################### BUTTERWORTH FILTERS ###########################

########## Low-Pass Filter ##########

ax1 = plt.subplot(221) # Select the subplot
b, a = sp.signal.butter(3, 0.5, 'low') # Compute the Butterworth filter
w, h = sp.signal.freqz(b, a) # Compute the frequency response
ax1.plot(w, 20 * np.log10(abs(h)), 'b') # Plot the response
ax1.set_xscale('log') # Modify the x-scale of the graph
ax1.set_title('Butterworth filter frequency response LOW-PASS') # Add title and labels
ax1.set_xlabel('Frequency [radians / second]')
ax1.set_ylabel('Amplitude [dB]')

########## High-Pass Filter ##########

ax3 = plt.subplot(223) # Select the subplot
b, a = sp.signal.butter(3, 0.1, 'high') # Compute the Butterworth filter
w, h = sp.signal.freqz(b, a) # Compute the frequency response
ax3.plot(w, 20 * np.log10(abs(h)), 'b') # Plot the response
ax3.set_xscale('log') # Modify the x-scale of the graph
ax3.set_title('Butterworth filter frequency response HIGH-PASS')  # Add title and labels
ax3.set_xlabel('Frequency [radians / second]')
ax3.set_ylabel('Amplitude [dB]')


########## Band-Pass Filter ##########

ax2 = plt.subplot(222) # Select the subplot
b, a = sp.signal.butter(3, [0.03, 0.5], 'bandpass') # Compute the Butterworth filter
w, h = sp.signal.freqz(b, a) # Compute the frequency response
ax2.plot(w, 20 * np.log10(abs(h)), 'b') # Plot the response
ax2.set_xscale('log') # Modify the x-scale of the graph
ax2.set_title('Butterworth filter frequency response BAND-PASS')  # Add title and labels
ax2.set_xlabel('Frequency [radians / second]')
ax2.set_ylabel('Amplitude [dB]')

########## Stop-Band Filter ##########

ax4 = plt.subplot(224) # Select the subplot
b, a = sp.signal.butter(3, [0.03, 0.5], 'bandstop') # Compute the Butterworth filter
w, h = sp.signal.freqz(b, a) # Compute the frequency response
ax4.plot(w, 20 * np.log10(abs(h)), 'b') # Plot the response
ax4.set_xscale('log') # Modify the x-scale of the graph
ax4.set_title('Butterworth filter frequency response STOP-BAND')  # Add title and labels
ax4.set_xlabel('Frequency [radians / second]')
ax4.set_ylabel('Amplitude [dB]')
```

```python
########################### CHEBYSHEV FILTERS ##########################

############ Low-Pass Filter ############

ax1 = plt.subplot(221) # Select the subplot
b, a = sp.signal.cheby1(3, 2, 0.5, 'low') # Compute the Chebyshev-I filter
w, h = sp.signal.freqz(b, a) # Compute the frequency response
ax1.plot(w, 20 * np.log10(abs(h)), 'b') # Plot the response
ax1.set_xscale('log') # Modify the x-scale of the graph
ax1.set_title('Chebyshev-I filter frequency response LOW-PASS') # Add title and labels
ax1.set_xlabel('Frequency [radians / second]')
ax1.set_ylabel('Amplitude [dB]')

############ High-Pass Filter ############

ax3 = plt.subplot(223) # Select the subplot
b, a = sp.signal.cheby1(3, 2, 0.1, 'high') # Compute the Chebyshev-I filter
w, h = sp.signal.freqz(b, a) # Compute the frequency response
ax3.plot(w, 20 * np.log10(abs(h)), 'b') # Plot the response
ax3.set_xscale('log') # Modify the x-scale of the graph
ax3.set_title('Chebyshev-I filter frequency response HIGH-PASS') # Add title and labels
ax3.set_xlabel('Frequency [radians / second]')
ax3.set_ylabel('Amplitude [dB]')

############ Band-Pass Filter ############

ax2 = plt.subplot(222) # Select the subplot
b, a = sp.signal.cheby1(3, 2, [0.03, 0.5], 'bandpass') # Compute the Chebyshev-I filter
w, h = sp.signal.freqz(b, a) # Compute the frequency response
ax2.plot(w, 20 * np.log10(abs(h)), 'b') # Plot the response
ax2.set_xscale('log') # Modify the x-scale of the graph
ax2.set_title('Chebyshev-I filter frequency response BAND-PASS') # Add title and labels
ax2.set_xlabel('Frequency [radians / second]')
ax2.set_ylabel('Amplitude [dB]')

############ Stop-Band Filter ############

ax4 = plt.subplot(224) # Select the subplot
b, a = sp.signal.cheby1(3, 2, [0.03, 0.5], 'bandstop') # Compute the Chebyshev-I filter
w, h = sp.signal.freqz(b, a) # Compute the frequency response
ax4.plot(w, 20 * np.log10(abs(h)), 'b') # Plot the response
ax4.set_xscale('log') # Modify the x-scale of the graph
ax4.set_title('Chebyshev-I filter frequency response STOP-BAND') # Add title and labels
ax4.set_xlabel('Frequency [radians / second]')
ax4.set_ylabel('Amplitude [dB]')
```

```python
for order in [3, 5, 10, 20]: # For all the order
    b, a = sp.signal.cheby1(order, 1, 0.05, 'low') # Compute the Chebyshev-I filter
    w, h = sp.signal.freqz(b, a) # Compute the frequency response
    plt.plot(w, 20 * np.log10(abs(h)), label="order = %d" % order) # Plot the response
plt.xscale('log') # Modify the x-scale of the graph
plt.title('Chebychev-I filter frequency response LOW-PASS') # Add title and labels
plt.xlabel('Frequency [radians / second]')
plt.ylabel('Amplitude [dB]')
plt.margins(0, 0.1) # Rescale the display
plt.grid(which='both', axis='both') # Add the grid
plt.legend(loc='best') # Add the legend
plt.show()
```

We see through this graph that until a point, when you increase the order, more your filter is powerfull (its decrease faster after the cutoff frequency). But after a point, a passing band become unstable ( it starts to filter this band).

```python
x = np.zeros(40) #Create a vector x 
x[19] = 1 #Transform the vector x into a dirac function (20)

plt.figure() #Plot the resulting signal and add the title/labels
plt.stem(x)
plt.title('Input function')
plt.xlabel('k')
plt.ylabel('X[k]=dirac(k-20)')
```

**2.2**

```python
T = 1 #Initialize T
s = 0.5
alpha = s * T #Compute alpha
a  = np.exp(-alpha) #Compute a

########################## CAUSAL PART ###########################

causalY = np.zeros(40*T) #Create a vector to store the results
for i in range(40*T): #Compute the causal Part of the filter
    causalY[i]= 2*a*causalY[i-1]-a*a*causalY[i-2]+x[i]+a*(alpha-1)*x[i-1]

plt.figure() #Plot the causal part and add the title/labels
plt.stem(causalY)
plt.title('Causal part of the smoothing filter')
plt.xlabel('k')
plt.ylabel('y[k]')

###################### ANTI-CAUSAL PART ########################

AcausalY = np.zeros(40*T) #Create a vector to store the results
for i in range(40*T-3, 0, -1): #Compute the anti-causal part
    AcausalY[i]= 2*a*AcausalY[i+1]-a*a*AcausalY[i+2]+a*(alpha+1)*x[i+1]-a*a*x[i+2]

plt.figure() #Plot the anti-causal part and add the title/labels
plt.stem(AcausalY)
plt.title('Anti - Causal part of the smoothing filter')
plt.xlabel('k')
plt.ylabel('y[k]')

####################### FULL FILTER ############################

Y = causalY + AcausalY #Compute the full filter

plt.figure() #Plot the resulting filter and add the title/labels
plt.stem(Y)
plt.title('Smoothing filter')
plt.xlabel('k')
plt.ylabel('y[k]')
```

```python
width = im.shape[0] #Get the height and the width from the image
height = im.shape[1]

T = 1 #Initialize T
s = 0.5
alpha = s * T #Compute alpha
a  = np.exp(-alpha) #Compute a

############################# FITER ALONG COLUMNS ##############################

Deriv_im_p1 = np.zeros(im.shape) #Create the array to store the causal part
for i in range (width): #Compute the causal part
    for j in range (height):
        Deriv_im_p1[i, j]= 2*a*Deriv_im_p1[i, j-1]-a*a*Deriv_im_p1[i, j-2]-s*alpha*a*im[i, j-1]

Deriv_im_p2 = np.zeros(im.shape) #Create the array to store the anti-causal part
for i in range (width): #Compute the anti-causal part
    for j in range (height-3, 0, -1):
        Deriv_im_p2[i, j]= 2*a*Deriv_im_p2[i, j+1]-a*a*Deriv_im_p2[i, j+2]+s*alpha*a*im[i, j+1]

Deriv_im = Deriv_im_p1+Deriv_im_p2 #Compute the full derivative filter

plt.figure() #Display the image
io.imshow(Deriv_im, cmap = cm.gray)
plt.title('Filter along columns')

########################## FILTER ALONG ROWS ###################################

Deriv_im2_p1 = np.zeros(im.shape) #Create the array to store the causal part
for j in range (height):
    for i in range (width):
        Deriv_im2_p1[i, j]= 2*a*Deriv_im2_p1[i-1, j]-a*a*Deriv_im2_p1[i-2, j]-s*alpha*a*im[i-1, j]

Deriv_im2_p2 = np.zeros(im.shape) #Create the array to store the anti-causal part
for j in range (height):
    for i in range (width-3, 0, -1):
        Deriv_im2_p2[i, j]= 2*a*Deriv_im2_p2[i+1, j]-a*a*Deriv_im2_p2[i+2, j]+s*alpha*a*im[i+1, j]

Deriv_im2 = Deriv_im2_p1+Deriv_im2_p2 #Compute the full derivative filter

plt.figure() #Display the image
io.imshow(Deriv_im2, cmap = cm.gray)
plt.title('Filter along rows')
```
